# Matplotlib

# Pie Chart

```python
import matplotlib.pyplot as plt
```

```python
# Data
m=[1,3,5,7,9]
# Explode the 1st wedge
e=(0.1,0,0,0.1,0)

plt.pie(m,explode=e)
plt.title('Pie chart')
plt.show()
```

# 3D Charts

```python
#Data 
x=[1,2,3,4,5]
y=[1,4,9,16,25]
z=[1,8,27,64,125]

# Create a figure object
fig=plt.figure()

#create the 3D plot
ax=plt.axes(projection='3d')
ax.plot3D(z,y,x)
```
