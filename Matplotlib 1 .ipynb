# Matplotlib
  - Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python.
  - It is particularly well-suited for creating plots and charts to visualize data.

# Import library

```python
#importing matplotlib module
from matplotlib import pyplot as plt
```

# line Plot

```python
# x-axis values
x=[5,4,6,2,9]
 
# y-axis values
y=[10,5,3,1,4]

#function to pl0t
plt.plot(x,y)
plt.show()
```

# Bar plot

```python
#function for barplot
plt.bar(x,y)
plt.show()
```

# Histogram

```python
#function for histogram
# funtion for X values
plt.hist(x)
plt.show

#function for y values
plt.hist(y)
plt.show()
```

# Scatter Plot

```python
# function for scatter plot
plt.scatter(x,y)
plt.show()
```

 # Labeling Axes and Title

```python
plt.scatter(x, y)
plt.title('Amir')
plt.xlabel('Wahid')
plt.ylabel('Akhtar')
plt.show()
```

# Multiple Graphs

```python
#scatter plot
plt.scatter(x,y)
plt.show()
#line graph
plt.plot(x,y)
plt.show()
# histogram
plt.hist(x)
plt.hist(y)
plt.show()
#box plot
plt.boxplot(x,y)
plt.show()
```

# Pie Chart

```python
#Data for Pie chart
name=["Akhtar","Amir","Wahid","Uzair"]
marks=[70,44,43,60]

#plot pie chart
plt.pie(marks,labels=name, autopct='%1.1f%%', startangle=90)
plt.title('Pie chart Example')
plt.show()
```

# Area Plot

```python
z=[4,7,9,2,4]

# Area chart
plt.fill_between(x, y, z, color='blue', alpha=0.4, label='Area 1-2')
plt.plot(x, y, label='Line 1', marker='o')
plt.plot(x, z, label='Line 2', marker='o')

plt.legend()
plt.show()
```
